```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## Load the dataset

```python
dataset = pd.read_csv('/kaggle/input/california-housing-prices/housing.csv')
dataset
```

```python
dataset.head(10)
```

# Data Preprocessing

## Display the statistical description of the dataset

```python
info = dataset.info()
info
```

# Data Visualization

## Plotting the data

```python
plt.scatter(dataset["median_income"], dataset["median_house_value"], color='blue')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.title('Relationship between Median Income and Median House Value')
plt.show()
```

# Linear Regression Analysis

## Splitting the dataset into training and test sets

```python
x = dataset["median_income"].values.reshape(-1, 1)
y = dataset["median_house_value"].values.reshape(-1, 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 1/3, random_state= 0)
```

## Plotting the training data

```python
plt.scatter(x_train, y_train, color= 'red')
plt.xlabel('median_income')
plt.ylabel('median_house_value')
plt.title('Training Data With Regression Line')
plt.show()
```

## Performing Linear Regression

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

## Show the regression model's score

```python
print("Regression Model Score:", regressor.score(x_test, y_test))
```

# Model Evaluation and Prediction

## Predicting the test set results

```python
y_pred = regressor.predict(x_test)
y_pred
```

## Plotting the regression line on the training data


```python
plt.scatter(x_train, y_train, color= 'red')
plt.plot(x_train, regressor.predict(x_train), color='blue')
plt.xlabel('median_income')
plt.ylabel('median_house_value')
plt.title('Training Data With Regression Line')
plt.show()
```

## Predict three different values from the dataset using the trained model

```python
predictions = regressor.predict(x_test[:3])
real_values = y_test[:3]
```

## Print the predicted and real values

```python
for i in range(3):
    print(f"Predicted value: {predictions[i][0]}")
    print(f"Real value from dataset: {real_values[i][0]}")
    print()
```
